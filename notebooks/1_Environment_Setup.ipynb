# Kubernetes Lab - Environment Setup

## Objective

Before deploying any applications, I needed a Kubernetes environment that I could use for the rest of the lab.

The goal of this section was to:

- Prepare three Ubuntu virtual machines
- Install Kubernetes components
- Configure containerd as the container runtime
- Initialize the Kubernetes control plane
- Join worker nodes to the cluster
- Verify that all nodes were communicating correctly

This environment is used throughout the remaining labs.
